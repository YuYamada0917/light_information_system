```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def plot_complex_amplitude_coil_with_axis():
    # 角度の範囲を定義 (0 から 8πまで)
    theta = np.linspace(0, 8 * np.pi, 1000)
    z = theta

    # 複素振幅の実部と虚部（サイン波とコサイン波）
    x = np.cos(theta)
    y = np.sin(theta)

    # 3Dプロットの設定
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x, y, z)

    # Z軸（実部と虚部が0の軸）をプロット
    ax.plot([0, 0], [0, 0], [0, max(z)], color='red', linestyle='--')

    ax.set_xlabel('Real Part')
    ax.set_ylabel('Imaginary Part')
    ax.set_zlabel('Axis')
    ax.set_title('Complex Amplitude Coil with Axis')
    plt.show()

plot_complex_amplitude_coil_with_axis()
```

```python
def plot_cosine_function():
    # 角度の範囲を定義 (0 から 8πまで)
    theta = np.linspace(0, 8* np.pi, 1000)

    cosine_values = np.cos(theta)

    # 2Dプロットの設定
    plt.figure()
    plt.plot(theta, cosine_values)

    plt.xlabel('Theta')
    plt.ylabel('Cosine Value')
    plt.title('Cosine Function Mapped from Real Part')

    plt.grid(True)
    plt.show()

plot_cosine_function()
```
